```python
import konlpy
from konlpy.tag import Kkma, Komoran, Hannanum, Okt
```

```python
# 불팔요한 문자 제거
import re

fixed_data = re.sub('[^가-힣]',' ', data)
fixed_data
```

```python
komoran = Komoran()

noun_data = komoran.nouns(fixed_data)
```

```python
len(noun_data)
```

```python
df_noun = pd.DataFrame({'noun' : noun_data})
df_noun
```

```python
df_noun['length'] = df_noun['noun'].str.len()
df_noun
```

```python
# 한 글자는 무의미하므로 두 글자 이상만 남기기
df_noun = df_noun.query('length >=2')
df_noun.sort_values('length')
```

```python
df_noun = df_noun.groupby('noun', as_index=False).agg(len= ('noun','count')).sort_values('len', ascending=False)
df_noun
```

```python
!pip install WordCloud
```

```python
# 데이터프레임 딕셔너리 변환
dic_word = df_noun.set_index('noun').to_dict()['len']
dic_word
```

```python
from wordcloud import WordCloud
wc = WordCloud(random_state = 42,
               font_path = '/content/drive/MyDrive/text/NanumSquareR.ttf',
               width = 400, # 가로크기
               height = 320, # 세로 크기
               background_color = 'black',
               colormap='inferno')
```

wordcloud 생성

```python
img_wordcloud = wc.generate_from_frequencies(dic_word)
```

```python
# 출력
plt.figure(figsize=(12,8))
plt.axis('off')           # 테두리 선 제거
plt.imshow(img_wordcloud)
```
